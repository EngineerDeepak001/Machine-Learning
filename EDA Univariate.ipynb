# EDA Using Univariate Analysis

## Har ek colums ko janna , unka analysis krna

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data 
## a. Countplot

```python
sns.countplot(df['Sex'])
# df['Sex'].value_counts().plot(kind='bar')
```

## b. PieChart

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

# 2. Numerical Data
  ## a. Histogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Age'], bins=50)
```

# b. Distplot

```python
sns.distplot(df['Age'])
```

# C. Boxplot 
   ## Five Numbers ki summary dena 

```python
sns.boxplot(df['Fare'])
```
