```python
#Cheacking by data ,heart dieseas is yes or not
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
 X= df.drop(columns=['target'])
y = df['target']
```

```python
X
```

```python
X.shape
```

y

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=8)
```

```python
print(X_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier  # Base model
from sklearn.neighbors import KNeighborsClassifier   # Base model
from sklearn.linear_model import LogisticRegression  # Meta model
from sklearn.ensemble import GradientBoostingClassifier # Base model
```

```python
estimators = [
    ('rf',RandomForestClassifier(n_estimators=10, random_state=42)),
    ('knn',KNeighborsClassifier(n_neighbors=10)),
    ('gbdt',GradientBoostingClassifier())
]
```

```python
from sklearn.ensemble import StackingClassifier

clf = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(),
    cv=10
)
```

# Train Model

```python

clf.fit(X_train, y_train)
```

# prediction

```python
y_pred = clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
